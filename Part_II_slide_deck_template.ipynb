> For this presentation, I focus on just the influence of user type on individual rides. I start by introducing the
duration variable, followed by the pattern in user type distribution, then the relationship betwwen user type and duration. 


## Dataset Overview

> This data contains information about individual rides made in a bike-sharing system covering the greater San Francisco Bay area in 2019. There are 2,407,259 individual rides in the dataset with 14 features. I feature engineered three new period variables and dropped 134,194 data points from my analysis to reduce negative bias.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
fileName = 'bike_sharing_clean.csv'
filePath = 'C:/datasets'
fullFileDir = filePath + '/' + fileName

bike = pd.read_csv(fullFileDir)
```

```python
# Convert start_time and end_time into datetime
timeCol = ['start_time', 'end_time']
for col in timeCol:
    bike[col] = pd.to_datetime(bike[col])
    
# Convert IDs from floats and ints to object data type
strCol = ['start_station_id', 'end_station_id', 'bike_id']
for col in strCol:
    bike[col] = bike[col].astype(str)
    
# Convert User Type to categorical data type
bike.user_type = bike.user_type.astype('category')
```

```python
# Convert time of day, day of week, and month into cartegorical type
bike['time_of_day'] = bike.time_of_day.astype('category')
bike['day_of_week'] = bike.day_of_week.astype('category')
bike['month'] = bike.month.astype('category')

# Convert day of week and into ordinal categorical type
dayOrder = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
monthOrder = ['January','February', 'March', 'April', 'May', 'June', 
              'July', 'August', 'September', 'October', 'November', 'December']

bike['day_of_week'] = bike['day_of_week'].cat.set_categories(dayOrder, ordered=True)
bike['month'] = bike['month'].cat.set_categories(monthOrder, ordered=True)
```

```python
# Create a function to polish plots with labels
def chartLabel(x,y,t):
    '''
    x = str(label on x axis)
    y = str(label on y axis)
    t = str(chart title)
    '''
    plt.title(t)
    plt.xlabel(x)
    plt.ylabel(y)
    plt.show()
```

## Distribution of Ride Duration

> Bike ride duration in the dataset take on a very large range of values, from about 60 secs at the lowest, to 80,000 secs at the highest. After removing outliers the distribution of duration takes on a unimodal shape and it's right skewed with a concentration of values between 250 and 500 seconds.

```python
plt.ticklabel_format(style='plain')

binEdges = np.arange(0, bike.duration_sec.max()+50, 50)
plt.hist(data=bike, x='duration_sec', bins=binEdges)
ticks = [0, 2e4, 4e4, 6e4, 8e4, 1e5, 12e4, 14e4, 16e4]
labels = ['0', '20k', '40k', '60k', '80k', '100k', '120k', '140k', '160k']
plt.yticks(ticks, labels)
plt.suptitle('Fig. 1');

chartLabel('Duration (sec)', 'Count', 'Distribution of Ride Duration')
```

## Who are the most active users of the scheme?

> User Type is a feature describing if a user is a registered member on the scheme or just a casual rider. A large proportion of trips on the bike sharing scheme were taken by subscribers.

```python
userTypeCount = bike.user_type.value_counts()
userLabel = bike.user_type.value_counts().index
plt.pie(userTypeCount, labels=userLabel, startangle=90, counterclock=False, wedgeprops={'width':0.5}, autopct='%1.1f%%')
plt.axis('square')
plt.suptitle('Fig. 2')

chartLabel('User Type', '', 'Proportion of Users of Bike Sharing Scheme')
```

## How does ride duration differ with user type?
Plotting ride duration by user type shows the difference in user behaviour between both groups. While registered members completed more trips lasting predominantly around 250 and 500 secs, casual riders showed a more even distribution with a fatter tail showing prevalence of long duration trips. 


```python
baseColor = sb.color_palette()[0]
ax = sb.violinplot(data=bike, x='user_type', y='duration_sec', color=baseColor, 
                   inner='quartile', order=['Subscriber', 'Customer']);
plt.suptitle('Fig. 3')
chartLabel('User Type', "Duration (sec)", 'Ride Duration by User Type')
```

## What's the driving the difference in duration by User Type?
There's a relationship visible between duration, user type, and variables of period. The plots below provide more insight into the drivers of the user type influence on ride duration. The shorter duration of Subscriber trips and the peak before the start of working hours, the decline in duration during official working hours and and the peak after that suggests that most subsribers are locals that rely heavily on the bike scheme to commute and get ahead in their respective busy lives. This proves that the scheme is serving its functional purpose. 


```python
# Create adaptation of clustered bar chart to clearly see relationship
ax = sb.boxplot(data=bike, x='day_of_week', y='duration_sec', hue='user_type');
plt.xticks(rotation=30)
plt.suptitle('Fig. 4.1')
plt.legend(bbox_to_anchor=(1.3, 0.75));
chartLabel('Day of Week', 'Avg. Ride Duration (sec)', 'Average Ride Duration by Day')
```

```python
sb.pointplot(data=bike, x='time_of_day', y='duration_sec', hue='user_type', dodge=0.3, ci='sd', size=5, linestyles='');
plt.legend(bbox_to_anchor=(1, 0.75))
plt.suptitle('Fig. 4.2');
chartLabel('Time of Day (Hour)', 'Avg. Ride Duration (sec)', 'Average Ride Duration by Hour')
```

## How Does User Activity Differ by Month?
The relationship with month doesn't show a clear contrasting pattern like the other period variables. Subscribers and Customers seem to maintain the same level of trip duration across months, with the longest trips in the spring. The significant drop in Customer ride duration in December might be connected to the effect of festive holidays coupled with the onset of winter.

```python
# Create adaptation of clustered bar chart to clearly see relationship
ax = sb.barplot(data=bike, x='month', y='duration_sec', hue='user_type', ci='sd');
plt.xticks(rotation=30)
plt.legend(bbox_to_anchor=(1.3, 0.75))
plt.suptitle('Fig. 5');

chartLabel('Month', 'Avg. Duration (Sec)', 'Average Ride Duration By Month')
```
